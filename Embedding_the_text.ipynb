```python
## Import Onehot from Keras Preprocessing text to produce Word indexes based on Vocabulary size
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## Sample Data
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
## One hot Representation of Sentences
one_hot_rep=[one_hot(word,10000) for word in sent] ## 10000 refers to vocabulary size
```

```python
one_hot_rep
```

```python
## Word Embedding . Layers for Embedding , Pad sequences for having all sentences with same shape
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
# Pad Sequences
sent_length=5
padded_sent=pad_sequences(one_hot_rep,maxlen=sent_length,padding='pre') ## Returns word index One hot rep in same shape
padded_sent
```

```python
## Embedding. Lets initialize Sequential for using Embedding
voc_size=10000
dim=10
model=Sequential()
model.add(Embedding(voc_size,dim))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
embedding_sent=model.predict(padded_sent)
```
