```python
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import pad_sequences
```

```python
# Load the pre trained model
model=load_model('simple_rnn_imdb.h5')
```

```python

model.summary()
```

## Prediction

Prediction for : Movie is not good but the screenplay is very good

```python
from tensorflow.keras.datasets import imdb
word_index=imdb.get_word_index()
```

```python
def preprocess(text):
  words=text.lower().split()
  encoded_text=[word_index.get(word,2)+3 for word in words]
  padded_text=pad_sequences([encoded_text],maxlen=500,padding='pre')
  return padded_text
```

```python
preprocess("Movie is not good but the screenplay is very good ")
```

```python
def prediction(review):
  preprocessed_review=preprocess(review)
  predicted_review=model.predict(preprocessed_review)
  sentiment= "Positive" if predicted_review[0][0]>0.8 else "Negative"
  return sentiment,predicted_review[0][0]
```

```python
# user Input and Output from Model Testing
text="This movie is so good but screenplay is bad and story is average"
sentiment,score= prediction("This movie is so good but screenplay is bad and story is average")

print(f"Prediction for {text} is : {sentiment} with {score}")
```
